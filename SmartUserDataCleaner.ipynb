```python
raw_data = "  AhMeT  yILMAZ ;  23 ;  1.78 ;  ahmetYILMAZ@GMAIL.Com   "
```

```python
raw_data = raw_data.strip()
isim, yas, boy, email = raw_data.split(";")
```

```python
isim = isim.lower().title().strip()
isim = isim[:5] + " " + isim[7:]
```

```python
yas = float(int(yas) + 10)
```

```python
boy = float(boy) * 100
```

```python
email = email.lower().strip()
kullaniciKodu = email[ : email.find("@")][: 3]
```

```python
print("Kullanıcı:", isim)
print("Yaş (10 yıl sonra):", yas)
print("Boy (cm):", boy)
print("Email Kullanıcı Kodu:", kullaniciKodu)
```
